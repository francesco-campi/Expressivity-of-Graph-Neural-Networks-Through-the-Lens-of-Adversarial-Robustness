# Read the results of the training

```python
import seml
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
nodes = 50
experiment_name = f'train_gin_er_{nodes}'
results: pd.DataFrame = seml.get_results(experiment_name, to_data_frame=True)
```

## Best results with respect to the MSE loss 

The models have been trained with the l1 loss and the best models has been selected wrt to the MSE loss

```python
results_mse = results[results["config.loss"] == "mse"]
substrucutres = results_mse['config.subgraph'].unique()
best_mse = pd.DataFrame(columns=["Best_model", "L1", "L1 std", "L1 count", "Mean", "Varience", "Std"], index=substrucutres)
for substrcuture in substrucutres:
    substructure_experiments = results_mse[results_mse['config.subgraph'] == substrcuture]
    best_experiment = substructure_experiments.sort_values(axis=0, by='result.l1_average').head(1).squeeze()
    best_mse.loc[substrcuture] = [best_experiment['_id'], best_experiment['result.l1_average'], best_experiment['result.l1_std_average'], 
    best_experiment['result.l1_count_average'], best_experiment['result.count_mean'], best_experiment['result.count_variance'], best_experiment['result.count_std']]
```

```python
font_dirs = ['/home/francesco/.local/share/fonts']
font_files = mpl.font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    mpl.font_manager.fontManager.addfont(font_file)

plt.rcParams['font.family'] = 'Linux Libertine'
plt.rcParams['font.size'] = 22

best_mse["Loss"] = 'MSE'
best_l1['Loss'] = 'L1'
best_l1_std['Loss'] = 'L1 std'
best_l1_count['Loss'] = 'L1 count'
results = pd.concat([best_mse, best_l1, best_l1_std, best_l1_count], axis=0)
results['Subgraph'] = results.index

metrics = ['L1', 'L1 std', 'L1 count']
fig, axes = plt.subplots(len(metrics), constrained_layout=True)
fig.set_figwidth(15)
fig.set_figheight(15)
fig.suptitle(f"GIN training results on Erdos-Reniy with {nodes} nodes", fontsize = 27)
for i, metric in enumerate(metrics):
    sns.barplot(data=results, x='Subgraph', y=metric, hue='Loss', ax=axes[i])
    axes[i].set_title(f'Errors w.r.t. the metric {metric}')
```
